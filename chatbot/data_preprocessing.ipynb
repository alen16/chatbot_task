```python
import pandas as pd
```

```python
file = "docs/assignment_articles.csv"
df = pd.read_csv(file)
df
```

```python
from bs4 import BeautifulSoup

cleanTexts = []
for row in df.itertuples(index=False):
    #print(i)
    title = row.title
    body = row.body
    print("Title: "+title+"\n")
    soup = BeautifulSoup(body)
    print("After BeautifulSoup:\n")
    print(soup.prettify())

    cleanText = ""
    for text in soup.find_all('p'):
        cleanText+= text.text+"\n"
    cleanTexts.append(cleanText)

print(len(cleanTexts))
```

```python
df['cleanTexts'] = cleanTexts
df
```

```python
import openai
import streamlit as st

openai.api_key = st.secrets["OPENAI_API_KEY"]

def get_cleand_text(html):
    full_response = ""
    for response in openai.ChatCompletion.create(
        model="gpt-4",
        messages=[
            {"role": "user", "content": "I'll give you an html. Can you extract clean text only in json format?"},
            {"role": "assistant", "content": "Of course! Please provide the HTML content you want me to process, and I'll extract the clean text from it and present it to you in JSON format."},
            {"role": "user", "content": html}
        ],
        stream=True,
    ):
        full_response += response.choices[0].delta.get("content", "")
    return full_response

response = get_cleand_text(html)
```

```python
import time
t=2 # 2 sec

cleanHtml = []
for row in df.itertuples(index=False):
    title = row.title
    body = row.body
    print("Title: "+title+"\n")
    response = get_cleand_text(body)
    cleanHtml.append(response)
    time.sleep(t)
```
