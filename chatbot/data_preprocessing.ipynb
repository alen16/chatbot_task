```python
from html.parser import HTMLParser
from html.entities import name2codepoint

class MyHTMLParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.texts = []
    
    def handle_starttag(self, tag, attrs):
        print("Start tag:", tag)
        for attr in attrs:
            print("     attr:", attr)

    def handle_endtag(self, tag):
        print("End tag  :", tag)

    def handle_data(self, data):
        clean_data = data.strip()
        if clean_data:  # Ignore whitespace strings
            self.texts.append(clean_data)

    def handle_comment(self, data):
        print("Comment  :", data)

    def handle_entityref(self, name):
        c = chr(name2codepoint[name])
        print("Named ent:", c)

    def handle_charref(self, name):
        if name.startswith('x'):
            c = chr(int(name[1:], 16))
        else:
            c = chr(int(name))
        print("Num ent  :", c)

    def handle_decl(self, data):
        print("Decl     :", data)

    def get_data(self):
        # Join all texts into one string
        return ' '.join(self.texts)
```

```python
body = df['body']

print(len(body))
```

```python
cleanTexts = []
for i in range(0, len(body)):
    parser = MyHTMLParser()
    parser.feed(body[i])
    text = parser.get_data()
    cleanTexts.append(text)
```

```python
df['cleanText'] = cleanTexts
df
```
