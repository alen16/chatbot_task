```python
import requests
import justext
import time
import socket

def fetch_content_by_url(url):
    full_text = ""
    fail = False
    try:
        response = requests.get(url)
        paragraphs = justext.justext(response.content, justext.get_stoplist("English"))
        
        for paragraph in paragraphs:
            if not paragraph.is_boilerplate:
                full_text+=" "+paragraph.text

    except:
        print("Connection time out with url: "+url)
        fail = True
    return full_text, fail
```

```python
text, fail = fetch_content_by_url(url)
print(text)
```

# NLU Crawler

```python
import requests
import json
import time

def get_taboola_nlu_response(service_url, article_url, headers = {'Accept-Encoding': 'gzip', 'Accept-Charset': 'utf-8'}):
    tStart = time.time() 
    
    r = requests.post(service_url, json={'urls': article_url, 'header': headers})
    tEnd = time.time()
    print('Cost %d seconds' % (tEnd - tStart))
    return json.dumps(r.json(), indent=2) 
```

```python
service_url = "http://newsroom-api-prod-cw.taboolasyndication.com/crawler/bulkText"
url = "https://www.journalinquirer.com/newsletters/hebron-school-switches-to-remote-learning-after-number-of-covid-19-cases/article_38d36aa2-10ca-11ec-b7d0-37317172d668.html"
result = get_taboola_nlu_response(service_url, url)
```

```python
print(result)
```

# Taboola Crawler - QA

```python
import requests
import time
def non_headless_crawler(url):
    tStart = time.time() 
    full_url = "http://newsroom-api-prod-cw.taboolasyndication.com/crawler/text?headless=false&url=" + url
    resp = requests.get(url=full_url)
    result = resp.json()
    tEnd = time.time()
    print('Cost %d seconds' % (tEnd - tStart))
    result = result["data"]["text"][0]
    print(result)
    if result['title'] is None:
        return {"title": None, "text": None}
    else:
        return {"title": result["title"], "status": result["status"], "cleanText": result["cleanText"][0:2000]}   
   
```

```python
article_result = non_headless_crawler(url)
print(article_result)
```

```python
print(article_result)
```
